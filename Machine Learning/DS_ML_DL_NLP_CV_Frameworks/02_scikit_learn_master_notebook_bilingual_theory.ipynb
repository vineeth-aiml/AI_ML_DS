# Part 1 — scikit-learn Philosophy and Core API


# 1. scikit-learn ante enti?

## English
scikit-learn is a general-purpose machine learning framework in Python for classical ML workflows. It provides a consistent API for preprocessing, model training, evaluation, model selection, pipelines, feature engineering support, and many supervised/unsupervised algorithms.

## Telugu
scikit-learn ante Python lo classical ML workflows kosam use ayye general-purpose machine learning framework. Preprocessing, model training, evaluation, model selection, pipelines, feature engineering support, mariyu supervised/unsupervised algorithms ki consistent API istundi.

## One-line meaning
**scikit-learn = classical ML workflow ki standard toolkit + framework.**


# 2. Framework philosophy

## Main idea
Different estimators and transformers ki oka common interface untundi.

## Why powerful?
Once you understand:
- fit
- transform
- fit_transform
- predict
- predict_proba
- score
- get_params
- set_params

almost framework usage easy avutundi.

## One-line meaning
**Learn one API pattern, use many models.**


# 3. Estimator ante enti?

## English
An estimator is any object in scikit-learn that can learn from data using `fit()`.

## Telugu
scikit-learn lo `fit()` method use chesi data nundi nerchukune object ni estimator antaru.

## Types
- classifier
- regressor
- clusterer
- transformer
- outlier detector
- meta-estimator

## One-line meaning
**Estimator = data nundi learn chese object.**


# 4. Transformer ante enti?

## English
A transformer changes input data into a different representation, usually via `fit`, `transform`, or `fit_transform`.

## Telugu
Transformer ante input data ni vere representation lo ki marache object. Usually `fit`, `transform`, `fit_transform` methods untayi.

## Examples
- StandardScaler
- OneHotEncoder
- PCA
- SimpleImputer

## One-line meaning
**Transformer = data representation ni marchese object.**


# 5. Core estimator methods

## fit(X, y=None)
Data nundi learn cheyyadam

## transform(X)
Data ni new representation lo ki marchadam

## fit_transform(X, y=None)
fit + transform shortcut

## predict(X)
Predictions ivvadam

## predict_proba(X)
Class probabilities ivvadam

## decision_function(X)
Raw decision scores

## score(X, y)
Default model score

## get_params()
Hyperparameters retrieve cheyyadam

## set_params()
Hyperparameters set/update cheyyadam

## One-line meaning
**scikit-learn API heart = fit / transform / predict pattern.**

# Part 8 — Practical Project Flow with scikit-learn


# 45. Real-world tabular ML flow in scikit-learn

## Typical steps
1. load data
2. split data
3. define numeric / categorical columns
4. build imputers + scalers + encoders
5. combine with ColumnTransformer
6. attach model using Pipeline
7. evaluate with CV
8. tune with GridSearchCV / RandomizedSearchCV
9. inspect metrics
10. analyze errors

## One-line meaning
**Best scikit-learn usage = pipeline-driven reproducible workflow.**


# 46. Why Pipeline + ColumnTransformer combo is so important

## Because real data lo
- missing values
- categories
- different numeric scales
untayi.

## Without pipeline
- leakage chance
- repeated code
- inconsistent preprocessing

## With pipeline
- clean workflow
- safe CV
- end-to-end tuning
- reproducibility better

## One-line meaning
**Pipeline + ColumnTransformer = real-world sklearn superpower.**


# 47. Common mistakes while using scikit-learn

- preprocessing before split
- LabelEncoder on feature columns blindly
- wrong metric choose cheyyadam
- default `score()` meeda over-rely avvadam
- no stratification in imbalanced data
- no pipeline use cheyyakapovadam
- wrong CV split
- feature leakage
- over-tuning with weak baseline

## One-line meaning
**scikit-learn easy anipinchina, workflow mistakes valla fake results ravachu.**
