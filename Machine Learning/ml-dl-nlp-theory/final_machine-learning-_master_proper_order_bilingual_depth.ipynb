# Part 1 — Foundations


# 1. Machine Learning ante enti? | What is Machine Learning?

## English
Machine Learning is the process of learning useful patterns from historical data and using those patterns to make predictions, classifications, rankings, scores, or decisions on new unseen data.

## Telugu
Machine Learning ante historical data nundi useful patterns nerchukuni, aa patterns ni use chesi kotha unseen data meeda predictions, classifications, rankings, scores, leka decisions cheyyadam.

## One-line meaning
**ML = learn from past data, then generalize to new data.**

## Intuition
Suppose 10,000 previous loan applications unnayi. Manual ga prathi case ki rule rayadam kastam. Kani income, credit history, defaults, repayment behavior ni model observe chesi risk pattern nerchukuntundi.

## Why it matters
ML valla:
- large-scale prediction cheyyachu
- dynamic patterns handle cheyyachu
- manual rules kanna adapt avvachu

## Interview-ready answer
Machine learning is useful when exact hand-written rules are difficult to define, but enough historical data exists from which a system can learn patterns and generalize those patterns to unseen examples.


# 2. ML enduku use chestham? | Why do we use ML?

## English
We use ML when the problem contains learnable patterns, the data is available, and manually writing robust rules is difficult or not scalable.

## Telugu
Problem lo learn cheyyadaniki patterns unnappudu, data available unnappudu, manual ga strong rules rayadam kastam leka scalable kaakapothe ML use chestham.

## One-line meaning
**Rules hard, data rich → ML useful.**

## When ML is useful
- prediction
- ranking
- personalization
- anomaly detection
- forecasting

## When ML may not be needed
- simple rule-based problem
- data very limited
- target itself unclear
- deterministic policy enough

## Interview-ready answer
I use ML only when the problem has measurable business value, enough data, and a pattern that cannot be captured reliably using simple manual rules.

# 5. Regression vs Classification

## Regression
Continuous numeric value predict chestundi.  
Examples: salary, sales, price  
**One-line:** Number predict chesthe regression.

## Classification
Class/category predict chestundi.  
Examples: spam / not spam, churn / no churn  
**One-line:** Class predict chesthe classification.

## Interview-ready answer
If the target is continuous, it is regression. If the target is categorical/discrete, it is classification.

# Part 3 — Learning Behavior and Model Generalization


# 13. Bias vs Variance

Bias ante oversimplified assumptions valla vacche error.  
Variance ante training data changes ki chala sensitive ga undadam valla vacche error.

## One-line meaning
**Bias = too simple, Variance = too unstable.**

## High bias indicators
- train low
- validation low

## High variance indicators
- train high
- validation much lower

## Reduce bias
- better model
- better features
- reduce excessive regularization

## Reduce variance
- simpler model
- more data
- regularization
- ensembles

## Interview-ready answer
Bias comes from oversimplification, while variance comes from over-sensitivity. I balance them using complexity control, features, data, and regularization.
