# 15) DS / Statistics concepts inka complete cheyyali

## 15.1 Variance asal enti?
Variance ante data points mean nundi entha spread ayi unnayo measure.

### Intuition
- variance low unte values mean daggara untayi
- variance high unte values chala spread ayi untayi

### Enduku square chestham?
- negative, positive deviations cancel kakudadhu
- bigger deviations ki bigger penalty ivvali

### One line
Variance = squared spread around mean.

---

## 15.2 Standard Deviation (STD) asal enti?
Standard deviation = variance ki square root.

### Enduku important?
Variance squared units lo untundi. STD original units lo untundi kabatti practical ga easy ga interpret cheyyachu.

### Example
Average delivery time = 30 mins
- STD = 2 mins → stable process
- STD = 20 mins → inconsistent process

### One line
STD = actual spread around mean in original units.

---

## 15.3 Variance vs Standard Deviation
- Variance = mathematical spread
- STD = practical spread
- Reporting, dashboards, interpretation lo STD ekkuva useful

### One line
Variance theoretical, STD practical.

---

## 15.4 Range asal enti?
Range = max - min

### Problem
Outliers unte range misleading avvachu.

### One line
Range = total spread, but outlier-sensitive.

---

## 15.5 Quartiles asal enti?
Quartiles dataset ni 4 equal parts ga divide chestayi.

- Q1 = 25% point
- Q2 = 50% point = median
- Q3 = 75% point

### Use cases
- salary distribution
- boxplots
- transaction spread
- skewed data summary

### One line
Quartiles = 4-part division points.

---

## 15.6 Percentiles asal enti?
Percentiles dataset ni 100 parts ga divide chestayi.

### Examples
- 90th percentile = 90% values aa point kanna takkuva
- 95th percentile = high-end values range

### Use cases
- exam rank
- salary benchmark
- P95 latency
- outlier threshold

### One line
Percentile = relative standing in data.

---

## 15.7 IQR (Interquartile Range) asal enti?
IQR = Q3 - Q1

Middle 50% data spread ni measure chestundi.

### Enduku important?
Outliers influence takkuva. Robust spread measure.

### Outlier rule
- Lower bound = Q1 - 1.5 * IQR
- Upper bound = Q3 + 1.5 * IQR

### One line
IQR = middle 50% spread.

---

## 15.8 Skewness asal enti?
Distribution left/right tilt ayi unda ani chupinche measure.

### Types
- Positive skew / right skew → tail right lo
- Negative skew / left skew → tail left lo
- Symmetric → mean ~ median

### Example
- salaries → usually right skew
- house prices → right skew
- easy exam marks → left skew avvachu

### One line
Skewness = distribution tilt direction.

---

## 15.9 Kurtosis asal enti?
Distribution tails entha heavy ga unnayo chupisthundi.

### Intuition
- high kurtosis → extreme values chance ekkuva
- low kurtosis → lighter tails

### One line
Kurtosis = tail heaviness / extremeness.

---

## 15.10 Covariance asal enti?
Rendu variables kalisi move avuthunnaya ani measure.

- positive covariance → okati perigithe inkoti kuda perugutundi
- negative covariance → okati perigithe inkoti tagguthundi

### Problem
Interpretation scale-dependent. Anduke correlation ekkuva use chestham.

### One line
Covariance = joint movement measure.

---

## 15.11 Correlation inka detail

### Types

### Values
- +1 strong positive

### Caution

### One line
Correlation = relation strength + direction.

## 15.13 Normal Distribution asal enti?
Bell-shaped, symmetric distribution.

### Properties
- mean = median = mode
- most values center daggara untayi
- extreme values rare

### 68-95-99.7 rule
- 68% → ±1 std
- 95% → ±2 std
- 99.7% → ±3 std

### One line
Normal distribution = ideal symmetric bell curve.

---

## 15.14 Z-score asal enti?
Value mean nundi enni standard deviations dooram lo undo.

Formula idea:
z = (x - mean) / std

### Use cases
- outlier detection
- scaling
- comparisons across different units

### Rule of thumb
|z| > 3 → possible outlier

### One line
Z-score = standardized distance from mean.

# 16) Python examples for important DS concepts

## 16.1 Mean, median, mode, variance, std
```python
import numpy as np
import pandas as pd

arr = np.array([10, 20, 20, 30, 40, 100])

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Variance:", np.var(arr))
print("Std:", np.std(arr))

s = pd.Series(arr)
print("Mode:", s.mode().tolist())
```

## 16.2 Quartiles, percentiles, IQR
```python
q1 = np.percentile(arr, 25)
q2 = np.percentile(arr, 50)
q3 = np.percentile(arr, 75)
iqr = q3 - q1

print("Q1:", q1)
print("Q2 / Median:", q2)
print("Q3:", q3)
print("IQR:", iqr)
```

## 16.3 Z-score
```python
mean = np.mean(arr)
std = np.std(arr)

z_scores = (arr - mean) / std
print(z_scores)
```

## 16.4 Correlation
```python
df = pd.DataFrame({
    "experience": [1, 2, 3, 4, 5],
    "salary": [3, 5, 7, 9, 11]
})

print(df.corr(numeric_only=True))
```

## 16.5 Missing values
```python
df = pd.DataFrame({
    "age": [23, 25, None, 29],
    "salary": [30000, None, 45000, 50000]
})

print(df.isnull().sum())
print(df.fillna(df.median(numeric_only=True)))
```

## 16.6 Standardization
```python
from sklearn.preprocessing import StandardScaler

X = [[20, 30000], [25, 40000], [30, 50000]]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

## 16.7 Normalization
```python
from sklearn.preprocessing import MinMaxScaler

X = [[20, 30000], [25, 40000], [30, 50000]]
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

# 21) Short interview-style answers

## Variance enti?
Mean nundi squared deviation average. Data spread entha undo chepthundi.

## Standard deviation enti?
Variance ki square root. Data mean nundi average spread ni original units lo chupisthundi.

## Quartile enti?
Sorted data ni 4 equal parts ga divide chese cutoff points.

## IQR enti?
Q3 - Q1. Middle 50% spread. Outlier detection lo useful.

## Percentile enti?
Oka value dataset lo relative ga ekkada undo chepthundi.

## Skewness enti?
Distribution left/right side ki tilt ayi unda ani chupisthundi.

## Z-score enti?
Value mean nundi enni standard deviations dooram lo undho chepthundi.

## Standardization vs normalization?
Standardization mean/std based, normalization min/max based.
