1) Variance asal enti?

Variance ante data points mean nundi entha dooram spread ayi unnayo measure chestundi.

Formula idea:

first mean calculate chestham
prati value mean nundi entha difference undo chustham
aa difference square chestham
average thesukuntam
Enduku square chestham?

Because:

negative, positive cancel kakudadhu
bigger deviations ki bigger penalty ivvali
Intuition

Variance pedda unte:

data chala spread ayi undi

Variance chinna unte:

values mean daggara close ga unnayi
Example

Data: 10, 12, 14
Mean = 12
Spread takkuva → variance takkuva

Data: 2, 12, 22
Mean = 12
Spread ekkuva → variance ekkuva

One line

Variance = data spread measure.


2) Standard Deviation (STD) asal enti?

Standard deviation = variance ki square root

Enduku use chestham?

Variance square units lo untundi.
Example:

salary rupees lo unte
variance rupees² lo ostundi
adi practical ga easy ga interpret cheyyalem

STD malli original units lo untundi.

Intuition
low std → data consistent
high std → data lo variation ekkuva
Real-time use cases
employee salaries spread
delivery times consistency
stock volatility
model error spread
anomaly detection
Example

Average delivery time = 30 mins

STD = 2 mins → stable process
STD = 20 mins → inconsistent process
One line

STD = average spread around mean in original units.


3) Variance vs Standard Deviation

Variance

squared spread
more mathematical

STD

easier to understand
real unit lo untundi
dashboards/reporting lo ekkuva use
One line

Variance = theoretical spread, STD = practical spread.


4) Quartiles asal enti?

Quartiles ante data ni 4 equal parts ga divide cheyyadam.

Sorted data lo:

Q1 = 25% point
Q2 = 50% point = median
Q3 = 75% point
Enduku use chestham?

Data distribution lo center and spread ni better ga ardham chesukodaniki.

Example

Marks sorted:
10, 20, 30, 40, 50, 60, 70, 80, 90

Q2 = 50
Q1 around lower 25% region
Q3 around upper 75% region
Use cases
salary distribution
house prices
transaction values
boxplot understanding
One line

Quartiles = dataset ni 4 parts ga split chese points.


5) Percentiles asal enti?

Quartiles 4 parts ayithe, percentiles 100 parts.

Meaning
90th percentile ante
aa value kanna 90% observations takkuva untayi
95th percentile ante top high values range
Use cases
exam ranking
salary benchmarking
latency metrics
outlier cutoffs
performance dashboards
Real example

App response time:

50th percentile = median latency
95th percentile = most users ki almost worst normal experience
99th percentile = extreme slow cases
One line

Percentile = ranking position in distribution.


6) IQR (Interquartile Range) asal enti?

IQR = Q3 - Q1

Middle 50% data spread ni chupisthundi.

Enduku use chestham?

Outliers impact lekunda spread measure cheyyadaniki.

Because mean/std outliers valla disturb avvachu.
Kani IQR robust.

Outlier rule

Usually:

lower bound = Q1 - 1.5 × IQR
upper bound = Q3 + 1.5 × IQR

Ee range bayata unna values outliers ga consider chestharu.

Use cases
boxplots
robust outlier detection
skewed data analysis
One line

IQR = middle 50% spread.


7) Range asal enti?

Range = max - min

Problem

Simple but too sensitive to extreme values.

Example:

salaries mostly 20k–40k
one CEO salary 5 lakh
range huge avuthundi

So range alone reliable kaadu.

One line

Range = total spread, but outlier-sensitive.


8) Skewness asal enti?

Skewness ante distribution left/right side ki tilt ayi unda ani.

Types
1. Right skew / positive skew

Tail right side lo untundi
Examples:

salaries
house prices
medical bills

Usually:
mean > median

2. Left skew / negative skew

Tail left side lo untundi
Examples:

easy exam marks where most scored high

Usually:
mean < median

3. Symmetric

Mean ≈ median

Why important?

Because:

mean misleading avvachu
transformations avasaram padachu
model assumptions break avvachu
One line

Skewness = distribution tilt direction.


9) Kurtosis asal enti?

Kurtosis distribution lo

tails entha heavy ga unnayi
extreme values chance entha undi
Intuition

High kurtosis:

more extreme outliers chance

Low kurtosis:

tails lighter
Practical

Interview lo deep maths kakapoyina,
just know:

outlier-prone distribution identify cheyyadaniki useful
One line

Kurtosis = tail heaviness / extremeness measure.


10) Covariance asal enti?

Covariance ante rendu variables kalisi ela move avuthunnayo measure.

Meaning
one perigithe ఇంకొటి kuda perigite → positive covariance
one perigithe ఇంకొటి taggite → negative covariance
Problem

Scale-dependent. Interpret cheyyadam easy kaadu.

Anduke correlation ekkuva use chestham.

One line

Covariance = two variables move together or opposite.

13) Normal Distribution asal enti?

Bell-shaped symmetric distribution.

Properties
mean = median = mode
most values center daggara untayi
extremes rare
68-95-99.7 rule

If data normal unte:

68% values mean ± 1 std
95% values mean ± 2 std
99.7% values mean ± 3 std
Use cases
z-scores
anomaly detection
statistical assumptions
confidence intervals approximation
One line

Normal distribution = ideal symmetric bell curve.


14) Z-score asal enti?

Z-score chepthundi:

Formula idea:

z=
std
x−mean
	​

Example

value = 70
z = 2
ante mean kanna 2 stds paina undi

Use cases

Outlier rough rule

|z| > 3 → possible outlier

One line

Z-score = value mean nundi relative distance.

20) Central Limit Theorem (CLT) asal enti?

Very important concept.

Simple ga:
Repeated ga samples thesukuni vati means chusthe,
original data normal kakapoyina,
sample means distribution normal laga behave chestundi
(large sample size unte).

Enduku important?

Because many statistical methods dani meeda depend avuthayi.

Use cases
confidence intervals
hypothesis testing
estimation
One line

CLT = sample means tend to become normal.


21) Standard Error asal enti?

Difference

One line

# 16) Python examples for important DS concepts

## 16.1 Mean, median, mode, variance, std
```python
import numpy as np
import pandas as pd

arr = np.array([10, 20, 20, 30, 40, 100])

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Variance:", np.var(arr))
print("Std:", np.std(arr))

s = pd.Series(arr)
print("Mode:", s.mode().tolist())
```

## 16.2 Quartiles, percentiles, IQR
```python
q1 = np.percentile(arr, 25)
q2 = np.percentile(arr, 50)
q3 = np.percentile(arr, 75)
iqr = q3 - q1

print("Q1:", q1)
print("Q2 / Median:", q2)
print("Q3:", q3)
print("IQR:", iqr)
```

## 16.3 Z-score
```python
mean = np.mean(arr)
std = np.std(arr)

z_scores = (arr - mean) / std
print(z_scores)
```

## 16.4 Correlation
```python
df = pd.DataFrame({
    "experience": [1, 2, 3, 4, 5],
    "salary": [3, 5, 7, 9, 11]
})

print(df.corr(numeric_only=True))
```

## 16.5 Missing values
```python
df = pd.DataFrame({
    "age": [23, 25, None, 29],
    "salary": [30000, None, 45000, 50000]
})

print(df.isnull().sum())
print(df.fillna(df.median(numeric_only=True)))
```

## 16.6 Standardization
```python
from sklearn.preprocessing import StandardScaler

X = [[20, 30000], [25, 40000], [30, 50000]]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

## 16.7 Normalization
```python
from sklearn.preprocessing import MinMaxScaler

X = [[20, 30000], [25, 40000], [30, 50000]]
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

# 21) Short interview-style answers

## Variance enti?
Mean nundi squared deviation average. Data spread entha undo chepthundi.

## Standard deviation enti?
Variance ki square root. Data mean nundi average spread ni original units lo chupisthundi.

## Quartile enti?
Sorted data ni 4 equal parts ga divide chese cutoff points.

## IQR enti?
Q3 - Q1. Middle 50% spread. Outlier detection lo useful.

## Percentile enti?
Oka value dataset lo relative ga ekkada undo chepthundi.

## Skewness enti?
Distribution left/right side ki tilt ayi unda ani chupisthundi.

## Z-score enti?
Value mean nundi enni standard deviations dooram lo undho chepthundi.

## Standardization vs normalization?
Standardization mean/std based, normalization min/max based.
