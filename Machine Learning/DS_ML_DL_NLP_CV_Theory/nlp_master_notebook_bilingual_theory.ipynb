# Part 1 — Foundations


# 1. NLP ante enti? | What is NLP?

## English
Natural Language Processing is the field of building systems that can understand, process, generate, analyze, or interact using human language.

## Telugu
Natural Language Processing ante human language ni understand cheyyagalige, process cheyyagalige, generate cheyyagalige, analyze cheyyagalige systems ni build chese field.

## One-line meaning
**NLP = computers ki language ardham ayyela cheyyadam.**

## Why important?
Human-generated data lo chala portion text form lo untundi:
- emails
- chats
- documents
- reviews
- support tickets
- search queries

## Interview-ready answer
NLP is the branch of AI that focuses on enabling machines to work with human language in ways such as classification, extraction, search, summarization, translation, and generation.


# 2. Text data enduku kastam? | Why is text hard?

## English
Text is difficult because language is ambiguous, context-dependent, noisy, and highly variable.

## Telugu
Text kastam endukante language ambiguous ga untundi, context meeda depend avuthundi, noisy ga untundi, chala variations untayi.

## One-line meaning
**Text simple ga kanipistundi, kani meaning complex ga untundi.**

## Why text is difficult
- same word ki multiple meanings
- same meaning ki multiple wordings
- sarcasm
- spelling mistakes
- abbreviations
- domain-specific vocabulary
- long-range dependencies

## Example
“bank” ante:
- river bank
- financial bank

Context lekapothe meaning clear kaadu.

# Part 2 — Classical NLP


# 8. Bag of Words (BoW)

## English
Bag of Words represents text by counting how many times each word appears, ignoring order.

## Telugu
Bag of Words text ni prathi word count basis meeda represent chestundi; word order ni ignore chestundi.

## One-line meaning
**Text = word counts.**

## Strengths
- simple
- interpretable

## Weaknesses
- order loss
- context loss
- sparse vectors


# 9. N-grams

## English
N-grams are contiguous sequences of N tokens.

## Telugu
N-grams ante continuous ga vacche N tokens sequence.

## Examples
- unigram
- bigram
- trigram

## One-line meaning
**N-grams = local word order capture cheyyadam.**


# 10. TF-IDF

## English
TF-IDF weighs words based on how important they are in a document relative to the whole corpus.

## Telugu
TF-IDF oka word document lo entha important undo, full corpus tho compare chesi weight istundi.

## One-line meaning
**Common-in-doc, rare-in-corpus words ki ekkuva importance.**

## Strengths
- strong classical baseline
- interpretable

## Weaknesses
- semantics weak
- context capture cheyyadu


# 11. Classical NLP representations limitations

## Limitations
- context weak
- synonyms separate ga treat chestundi
- ambiguity handle kastam
- sparse features
- long dependency understanding weak

## One-line meaning
**Classical features useful but semantics limited.**


# Part 3 — Representation Learning


# 12. Word Embeddings ante enti?

## English
Word embeddings are dense vector representations of words where semantically similar words tend to have similar vectors.

## Telugu
Word embeddings ante words ni dense vectors ga represent chese method. Semantically similar words ki similar vectors untayi.

## One-line meaning
**Word meaning ni vector form lo represent cheyyadam.**


# 13. Word2Vec

## English
Word2Vec learns embeddings by predicting context from words or words from context.

## Telugu
Word2Vec context ni batti word predict cheyyadam leka word ni batti context predict cheyyadam dvara embeddings nerchukuntundi.

## Two main approaches
- CBOW
- Skip-gram

## One-line meaning
**Word2Vec = context-based word meaning learning.**


# 14. GloVe

## English
GloVe learns embeddings using global word co-occurrence statistics from the corpus.

## Telugu
GloVe full corpus word co-occurrence statistics ni use chesi embeddings nerchukuntundi.

## One-line meaning
**Global co-occurrence based embedding method.**


# 15. FastText

## English
FastText represents words using character n-grams, helping with rare words and subword information.

## Telugu
FastText words ni character n-grams use chesi represent chestundi. Rare words mariyu subword information handle cheyyadaniki help chestundi.

## One-line meaning
**FastText = word + subword aware embeddings.**


# 16. Static vs Contextual Embeddings

## Static embeddings
One word ki one vector

## Contextual embeddings
Context batti same word ki different vector

## One-line meaning
**Static = one word, one vector; Contextual = one word, many context-based vectors.**
