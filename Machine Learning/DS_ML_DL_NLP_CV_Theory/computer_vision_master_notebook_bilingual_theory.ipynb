## Why difficult?
- lighting changes
- object rotation
- scale variation
- background clutter
- occlusion
- blur
- noise
- camera angle
- viewpoint changes
- inter-class similarity
- intra-class variation

## Example

- close-up
- side view
- night image

- motion blur

## Interview-ready answer
Computer vision is challenging because the visual appearance of the same semantic object can change drastically with environmental and geometric conditions, while different objects may also appear visually similar.
