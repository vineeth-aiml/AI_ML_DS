# 5. Spread / variability concepts

## 5.1 Variance
**Variance = mean nundi squared spread**

### Formula idea
- mean calculate chestham
- prati value mean nundi distance chustham
- aa distance square chestham
- average thesukuntam

### Enduku square chestham?
- negative and positive cancel kakudadhu
- bigger deviations ki bigger penalty ivvali

### Intuition
- variance low → data close ga undi
- variance high → data spread ayi undi

### One line
**Variance = data spread measure.**

---

## 5.2 Standard Deviation (STD)
**Standard deviation = variance ki square root**

### Enduku use chestham?
Variance squared units lo untundi. STD original units lo untundi, kabatti easy ga interpret cheyyachu.

### Example
Average delivery time = 30 mins  
- STD = 2 mins → stable process  
- STD = 20 mins → inconsistent process

### Real use cases
- employee salary variation
- process stability
- stock volatility
- anomaly detection
- model error spread

### One line
**STD = average spread around mean in original units.**

---

## 5.3 Variance vs Standard Deviation
- Variance → more mathematical
- STD → easier to understand
- Reporting lo STD ekkuva useful

### One line
**Variance = theoretical spread, STD = practical spread.**

---

## 5.4 Range
**Range = max - min**

### Problem
Simple but outlier-sensitive.

### One line
**Range = total spread, but extreme values valla distort avvachu.**


# 6. Quartiles, Percentiles, IQR

## 6.1 Quartiles
Dataset ni **4 equal parts** ga divide chese points.

- **Q1** = 25% point
- **Q2** = 50% point = median
- **Q3** = 75% point

### Enduku useful?
Distribution lo center and spread ni better ga ardham chesukodaniki.

### Use cases
- salary distribution
- house prices
- boxplot understanding
- transaction values

### One line
**Quartiles = dataset ni 4 parts ga split chese points.**

---

## 6.2 Percentiles
Quartiles 4 parts ayithe, **percentiles 100 parts**.

### Meaning
- 90th percentile → 90% values aa point kanna takkuva untayi
- 95th percentile → top high range

### Use cases
- exam rankings
- salary benchmarking
- latency metrics
- performance dashboards
- outlier cutoffs

### One line
**Percentile = ranking position in distribution.**

---

## 6.3 IQR (Interquartile Range)
**IQR = Q3 - Q1**

Middle 50% data spread ni chupisthundi.

### Enduku important?
Outliers impact takkuva. Robust spread measure.

### Outlier rule
- lower bound = `Q1 - 1.5 × IQR`
- upper bound = `Q3 + 1.5 × IQR`

Ee range bayata unna values ni usually outliers ga consider chestharu.

### One line
**IQR = middle 50% spread.**

## 7.2 Skewness
Distribution left/right side ki tilt ayi unda ani chupisthundi.

### Types
#### Right skew / positive skew
- tail right side lo untundi
- mean > median
- examples: salaries, house prices, medical bills

#### Left skew / negative skew
- tail left side lo untundi
- mean < median
- example: easy exam marks

#### Symmetric
- mean ≈ median

### Why important?
- mean misleading avvachu
- transformations avasaram padachu
- model assumptions break avvachu

### One line
**Skewness = distribution tilt direction.**

---

## 7.3 Kurtosis
Distribution tails entha heavy ga unnayo chupisthundi.

### Intuition
- high kurtosis → extreme outliers chance ekkuva
- low kurtosis → tails lighter

### One line
**Kurtosis = tail heaviness / extremeness measure.**

# 12. Normal distribution & Z-score

## 12.1 Normal Distribution
Bell-shaped symmetric distribution.

### Properties
- mean = median = mode
- most values center daggara untayi
- extremes rare

### 68-95-99.7 rule
If data normal unte:
- 68% values mean ± 1 std
- 95% values mean ± 2 std
- 99.7% values mean ± 3 std

### Use cases
- z-score
- anomaly detection
- statistical assumptions
- confidence intervals approximation

### One line
**Normal distribution = ideal symmetric bell curve.**

---

## 12.2 Z-score
Oka value mean nundi enni standard deviations dooram lo undho chepthundi.

### Formula
\[
z = rac{x - mean}{std}
\]

### Example
mean = 50, std = 10, value = 70  
z = 2

### Use cases
- outlier detection
- standardization
- comparison across scales

### Rough rule
`|z| > 3` → possible outlier

### One line
**Z-score = mean nundi standardized distance.**

# 21. Python examples for important DS concepts

## 21.1 Mean, median, mode, variance, std
```python
import numpy as np
import pandas as pd

arr = np.array([10, 20, 20, 30, 40, 100])

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Variance:", np.var(arr))
print("Std:", np.std(arr))

s = pd.Series(arr)
print("Mode:", s.mode().tolist())
```

## 21.2 Quartiles, percentiles, IQR
```python
q1 = np.percentile(arr, 25)
q2 = np.percentile(arr, 50)
q3 = np.percentile(arr, 75)
iqr = q3 - q1

print("Q1:", q1)
print("Q2 / Median:", q2)
print("Q3:", q3)
print("IQR:", iqr)
```

## 21.3 Z-score
```python
mean = np.mean(arr)
std = np.std(arr)

z_scores = (arr - mean) / std
print(z_scores)
```

## 21.4 Correlation
```python
df = pd.DataFrame({
    "experience": [1, 2, 3, 4, 5],
    "salary": [3, 5, 7, 9, 11]
})

print(df.corr(numeric_only=True))
```

## 21.5 Missing values
```python
df = pd.DataFrame({
    "age": [23, 25, None, 29],
    "salary": [30000, None, 45000, 50000]
})

print(df.isnull().sum())
print(df.fillna(df.median(numeric_only=True)))
```

## 21.6 Standardization
```python
from sklearn.preprocessing import StandardScaler

X = [[20, 30000], [25, 40000], [30, 50000]]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

## 21.7 Normalization
```python
from sklearn.preprocessing import MinMaxScaler

X = [[20, 30000], [25, 40000], [30, 50000]]
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```
