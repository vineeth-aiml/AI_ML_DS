
## 1) EDA ante asal enti?
EDA = Exploratory Data Analysis
Simple meaning
Dataset ni direct ga model ki ivvakunda,
mundu adi ela undi ani investigate cheyyadam.
Enduku use chestham?

* Because raw data mostly:

dirty ga untundi

missing values untayi

duplicates untayi

outliers untayi

wrong types untayi

skewed distributions untayi

useless columns untayi

leakage untadi

target imbalance untadi

##### EDA cheyakapothe emi avuthundi?

wrong model build chesthav

wrong features use chesthav

wrong cleaning chesthav

business ki wrong story chepthav

model offline lo baaga work ayi, production lo fail avvachu

##### One line
EDA ante data ni maatladinche process.

## 2) NumPy asal enti?
NumPy = Numerical Python

Asal enti?
Python lo numbers, arrays, matrix operations fast ga handle cheyyadaniki use ayye library.

Enduku use chestham?
Python list tho kuda numbers handle cheyyachu.
Kani big data/math operations kosam slow untundi.

##### NumPy:
fast

vectorized

memory efficient

math/stat operations easy

Ekkada use chestham?

arrays create cheyyadaniki

mean, std, percentile

matrix calculations

broadcasting

log transforms

clipping

random number generation

simulations

model input preparation


##### Example use

import numpy as np

arr = np.array([10, 20, 30, 40])

print(np.mean(arr))

print(np.std(arr))


##### Real project lo use

Suppose house price dataset undi:

price distribution skew undi

log transform cheyyali

z-score calculate cheyyali

percentiles chudali

Akkada NumPy useful.

##### Use cheyakapothe?

loops ekkuva rayali

code slow untundi

array math messy avuthundi

#### One line
NumPy = fast mathematical engine.

## 7) Statistics asal enti?

Statistics = data behavior ni measure cheyyadam

#### Asal enti?

Data lo:

center ekkada undi

spread entha undi

outliers unnaya

relation unda

sample nundi conclusion cheyyacha ani ardham chesukovadaniki use ayye science.

* mean, median, variance, standard deviation, outliers, hypothesis testing, confidence intervals anni model mundu important ani. 

## 7.1 Mean enduku use chestham?

##### Asal enti?

Average.

##### Enduku use chestham?

Center roughly ekkada undi telustundi.

Ekkada use chestham?

average salary

average sales

average trip duration

average marks


#### Problem enti?

Outliers unte misleading.

#### Example
10, 20, 30, 1000 → mean = 265

Kani actual typical value kaadu.

#### One line
Mean = average, but outlier-sensitive.

## 7.2 Median enduku use chestham?
##### Asal enti?
Middle value.

##### Enduku use chestham?

Skewed data lo actual typical center chupisthundi.

#### Ekkada use chestham?

house prices

salaries

medical bills

order values

#### Why powerful?

Outlier valla disturb avvadu.

#### One line
Median = robust center.

## 7.3 Mode enduku use chestham?

#### Asal enti?

Most frequent value.

Enduku use chestham?

Most common category telusukovadaniki.

#### Ekkada use chestham?

most common payment method

most common city

most common issue type

most common product category

#### One line
Mode = most common behavior.

## 7.4 Variance / Standard Deviation enduku use chestham?

##### Asal enti?

Values mean daggara clustered unnaya leka spread out ayyaya ani measure chestayi.

Enduku use chestham?

Same mean unna datasets kuda different spread kaligi undachu.

##### Ekkada use chestham?

process stability

delivery time consistency

risk

anomaly detection

z-score normalization

#### One line
Mean where data sits ani chepthundi; std how spread out it is ani chepthundi.

## 7.5 Outliers enduku identify chestham?
#### Asal enti?

Chala extreme values.

#### Enduku use chestham?
* Because outliers:

mean ni distort chestayi

variance inflate chestayi

regression ni bend chestayi

clusters ni break chestayi

Kani always remove chesthama? No.

First ask

data error aa?

fraud aa?

anomaly aa?

VIP case aa?

rare but valid aa?

#### One line
Outlier = rare point, sometimes problem, sometimes real business signal.


## 7.6 Correlation enduku use chestham?
##### Asal enti?

Rendu variables kalisi move avuthunnaya ani measure.

##### Enduku use chestham?

Feature relationships ardham avuthayi.

##### Ekkada use chestham?

feature selection

multicollinearity

exploratory analysis

Caution

Correlation ≠ causation

#### One line
Correlation relation chupisthundi, reason kaadu.


## 7.7 Hypothesis testing enduku use chestham?

##### Enduku use chestham?

### Ekkada use chestham?

#### One line
Hypothesis testing = signal vs random noise check.

## 12) Project lo exact order enti?

Before model building proper workflow

Step 1: Load data

Enduku? Raw structure teliyali

Step 2: Check shape, columns, dtypes

Enduku? Wrong data types catch cheyyali

Step 3: Missing values

Enduku? Null behavior ardham chesukovali

Step 4: Duplicates

Enduku? Double records avoid cheyyali

Step 5: Univariate analysis

Enduku? Each feature behavior teliyali

Step 6: Bivariate analysis

Enduku? Relations chudali

Step 7: Outlier analysis

Enduku? Extreme points impact ardham chesukovali

Step 8: Statistical testing

Enduku? Observed differences real aa ani verify cheyyali

Step 9: Save cleaned CSV

Enduku? Modeling-ready data prepare cheyyali

Step 10: Then only model

Enduku? Leka pothhe blind modeling avuthundi
