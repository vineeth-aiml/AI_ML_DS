📊 PART 6 — DATA AGGREGATION & GROUPING

🔟 GroupBy Operations

Definition: Splitting data into groups and computing aggregate stats.
    
When: For summaries or segment-based analysis.
    
Why: To find insights per group (e.g., avg salary by department).

```python
import pandas as pd

# -------------------------------------------
# 🔹 Step 1: Create sample employee dataset
# -------------------------------------------
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan', 'Fiona'],
    'Gender': ['F', 'M', 'M', 'F', 'M', 'F'],
    'Department': ['HR', 'IT', 'IT', 'HR', 'Finance', 'Finance'],
    'Salary': [50000, 60000, 55000, 52000, 70000, 68000]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# -------------------------------------------
# 🔹 Step 2: Basic GroupBy — Mean Salary by Department
# -------------------------------------------
dept_mean = df.groupby('Department')['Salary'].mean()
print("\nMean Salary by Department:")
print(dept_mean)

# -------------------------------------------
# 🔹 Step 3: Advanced GroupBy — Multiple Columns & Aggregations
# -------------------------------------------
dept_gender_stats = df.groupby(['Department', 'Gender']).agg({
    'Salary': ['mean', 'max']
})
print("\nSalary Stats by Department & Gender:")
print(dept_gender_stats)

# -------------------------------------------
# 🔹 Step 4: Final DataFrame
# -------------------------------------------
print("\nFinal DataFrame:")
print(df)
```

| Concept                  | Function                                         | Use                                                |
| ------------------------ | ------------------------------------------------ | -------------------------------------------------- |
| **Single groupby**       | `df.groupby('Department')['Salary'].mean()`      | Calculate average salary per department            |
| **Multi-column groupby** | `df.groupby(['Department','Gender']).agg({...})` | Perform multiple aggregations by multiple columns  |
| **agg()**                | Combine different metrics (`mean`, `max`, etc.)  | For richer summaries like average + highest salary |


1️⃣1️⃣ Pivot Tables & Crosstab

Definition: Excel-like summary table by category.
    
When: When you need summarized metrics across two dimensions.
    
Why: Easy for reporting and dashboards.


```python

import pandas as pd

# -------------------------------------------
# Step 1: Create sample sales data
# -------------------------------------------
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Year': [2023, 2023, 2023, 2023, 2024, 2024, 2024, 2024],
    'Sales': [12000, 10000, 8000, 9000, 15000, 11000, 9500, 10500]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# -------------------------------------------
# Step 2: Pivot Table — Summarize Sales by Region & Year
# -------------------------------------------
pivot = pd.pivot_table(df, values='Sales', index='Region', columns='Year', aggfunc='sum')
print("\nPivot Table (Total Sales by Region & Year):")
print(pivot)

# -------------------------------------------
# Step 3: Optional — Crosstab (alternative view)
# -------------------------------------------
crosstab = pd.crosstab(df['Region'], df['Year'], values=df['Sales'], aggfunc='sum')
print("\nCrosstab (Alternative Summary):")
print(crosstab)
```

🔧 PART 7 — DATA MERGING & JOINING

Definition: Combining multiple DataFrames into one.
    
When: When data is spread across different files/tables.
    
Why: To bring related information together.

```python
import pandas as pd

# -------------------------------------------
# Step 1: Create two DataFrames
# -------------------------------------------
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 28, 30]
})

df2 = pd.DataFrame({
    'id': [3, 4, 5],
    'Department': ['HR', 'Finance', 'IT'],
    'Salary': [60000, 65000, 70000]
})

print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

# -------------------------------------------
# Step 2: Concatenate (stack vertically)
# -------------------------------------------
concat_df = pd.concat([df1, df2], ignore_index=True)
print("\nConcatenated DataFrame (Vertical Stack):")
print(concat_df)

# -------------------------------------------
# Step 3: Merge (SQL-style join on 'id')
# -------------------------------------------
merged_df = pd.merge(df1, df2, on='id', how='inner')
print("\nMerged DataFrame (Inner Join on 'id'):")
print(merged_df)

# -------------------------------------------
# Step 4: Join (by index)
# -------------------------------------------
# Reindex df2 for demonstration
df2_indexed = df2.set_index('id')
df1_indexed = df1.set_index('id')

joined_df = df1_indexed.join(df2_indexed, lsuffix='_A', rsuffix='_B')
print("\nJoined DataFrame (by index):")
print(joined_df)
```

✅ Summary:

concat() → stacking

merge() → relational join

join() → by index

🧠 PART 9 — TIME SERIES ANALYSIS

Definition: 
Time Series Analysis is working with date- or time-indexed data — analyzing trends, periodicity, and seasonality.
    
📅 When

Use it when your dataset involves temporal patterns, such as:

Stock market prices 📈

Monthly sales 📊

Daily temperatures 🌡️

IoT or sensor readings ⏱️.
    
💡 Why

Enables resampling (aggregate by day, month, year)

Allows shifting (lag/lead for trend comparison)

Simplifies time-based filtering & rolling analysis

```python
import pandas as pd

# Sample daily sales data
data = {
    'Date': [
        '2024-01-01', '2024-01-02', '2024-01-03',
        '2024-02-01', '2024-02-02', '2024-02-03',
        '2024-03-01', '2024-03-02'
    ],
    'Sales': [200, 220, 210, 250, 270, 260, 300, 320]
}

df = pd.DataFrame(data)
print("Original Data:")
print(df)

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Set Date as index for time-based operations
df.set_index('Date', inplace=True)

# Resample monthly — aggregate total sales by month
monthly_sales = df.resample('M').sum()

print("\nMonthly Sales (Resampled):")
print(monthly_sales)
```

| Method                             | Purpose                                        |
| ---------------------------------- | ---------------------------------------------- |
| `df.shift(1)`                      | Shift values down by 1 day/month (lag feature) |
| `df.resample('W').mean()`          | Weekly average                                 |
| `df['Sales'].rolling('7D').mean()` | 7-day rolling average                          |
| `df.loc['2024-02']`                | Filter all February data                       |


```python
import pandas as pd

# ------------------------------
# 📅 Create Sample Time Series Data
# ------------------------------
data = {
    'Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'Sales': [200, 210, 220, 250, 260, 270, 300, 310, 320, 330]
}

df = pd.DataFrame(data)
df.set_index('Date', inplace=True)
print("Original Data:\n", df, "\n")

# ------------------------------
# 1️⃣ Shift: Lag previous day's sales
# ------------------------------
df['Prev_Day_Sales'] = df['Sales'].shift(1)
print("After Shift (Lag Feature):\n", df, "\n")

# ------------------------------
# 2️⃣ Resample Weekly: Compute weekly mean
# ------------------------------
weekly_avg = df.resample('W').mean()
print("Weekly Average Sales:\n", weekly_avg, "\n")

# ------------------------------
# 3️⃣ Rolling Window: 7-day rolling average
# ------------------------------
df['7D_Rolling_Avg'] = df['Sales'].rolling('7D').mean()
print("7-Day Rolling Average:\n", df, "\n")

# ------------------------------
# 4️⃣ Filter Data: February only
# ------------------------------
feb_data = df.loc['2024-01-05':'2024-01-10']
print("Filtered (Custom Range Example ~ Feb-like slice):\n", feb_data, "\n")
```

4️⃣ Apply Custom Functions

Definition: Apply your own logic to transform columns or entire DataFrames.
    
When: You need transformations not covered by built-ins.
    
Why: Flexibility for custom preprocessing.

```python
import pandas as pd

df = pd.DataFrame({'Score': [100, 27, 65, 75, 99]})

def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

df['Normalized'] = normalize(df['Score'])
print(df)
```

| Function      | Scope                    | Example                |
| ------------- | ------------------------ | ---------------------- |
| `.apply()`    | Column-wise              | df['x'].apply(func)    |
| `.applymap()` | Element-wise             | df.applymap(str.upper) |
| `.map()`      | Series element transform | df['col'].map(func)    |


```python
import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': ['x', 'y', 'z']
})

# 1️⃣ apply() → column-wise
df['A_doubled'] = df['A'].apply(lambda x: x * 2)
print(df.head())
print()
# 2️⃣ applymap() → element-wise (for entire DataFrame)
df_str = df[['B']].applymap(str.upper)

print("\napplymap example:\n", df_str)
print()
# 3️⃣ map() → element transform (Series only)
df['B_mapped'] = df['B'].map({'x': 'X-ray', 'y': 'Yellow', 'z': 'Zebra'})

print(df)
```
